# Pandas groupby is a function for grouping data objects into series (columns) or DataFrames(a group of series) based on particular indicator. In simpler terms, group by in python makes the management of datasets easier since you can put related records into groups

```python
import numpy as np
import pandas as pd
data = {
    "Employee": ["Alice", "Bob", "Charlie", "David", "Emma", "Charlie", "Grace", "Hannah"],
    "City": ["New York", "Los Angeles", "Chicago", "New York", "Chicago", "Philadelphia", "San Antonio", "San Diego"],
    "Product": [101, 102, 103, 104, 105, 5, 10, 108],
    "Hours": [35, 40, 30, 45, 25, 38, 42, 36]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby(['Employee']).sum()
```

```python
# count how many employee are from same city
df.groupby(['City']).count()
```

```python
# Sort based on hours
df.groupby('Employee')['Hours'].sum().reset_index().sort_values(by='Hours')
```
