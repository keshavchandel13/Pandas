```python
# a pivot table as a way to summarize and reorganize your data to see it from different perspectives.
pd.pivot_table(df,index=['Manager', 'Rep'], values=['Price'])
```

```python
# aggfunc: specifies the aggregation function to be used when summarizing the data in the pivot table.
pd.pivot_table(df,index=['Manager', 'Rep'], values=['Price'],aggfunc=[np.mean,len])
```

```python
finalTable=pd.pivot_table(df, index=['Manager', 'Status'],columns=['Product'],values=['Quantity', 'Price'], aggfunc={'Quantity': len, 'Price':[np.sum, np.mean]}, fill_value=0)
```
